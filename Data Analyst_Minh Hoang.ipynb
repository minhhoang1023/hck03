### Part 1: Data preprocessing

Firstly, important libraries were imported to perform data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns


%matplotlib inline
```

```python
features = pd.read_csv('features.csv', sep = '|')
his_sales = pd.read_csv('historic_sales.csv', sep = '|')
stores = pd.read_csv('store_info.csv', sep = ';')
```

```python
features = features.drop(columns='Unnamed: 0')
his_sales = his_sales.drop(columns='Unnamed: 0')
stores = stores.drop(columns='Unnamed: 0')
```

The Date column was then converted to datetime format

```python
features['Date'] = pd.to_datetime(features['Date'], yearfirst='True',infer_datetime_format=True, errors='coerce')
his_sales['Date'] = pd.to_datetime(his_sales['Date'], yearfirst='True',infer_datetime_format=True, errors='coerce')
```

```python
features.dtypes
```

```python
his_sales.dtypes
```

We can see that the "IsHoliday" column is in two different data types. That would not allow us to merge the dataframes. So let's first convert them to the same data type

```python
his_sales['IsHoliday'] = his_sales['IsHoliday'].astype(bool)
df1 = pd.merge(his_sales, stores,on='Store')
df1 = pd.merge(df1, features, on=['Store', 'Date', 'IsHoliday'])
df1.head(5)
```

Before having a deep dive into the data, let's make sure all columns are in proper data types

```python
df1.dtypes
```

Weekly_Sales column and Dept are currently in the wrong formats so let's fix that. We also need to encode the Type column.

```python
df1['Weekly_Sales'] = df1['Weekly_Sales'].str.replace(',','.')
df1['Weekly_Sales'] = df1['Weekly_Sales'].str.strip()
df1['Weekly_Sales'] = df1['Weekly_Sales'].astype(float)

df1['Dept'] = df1['Dept'].astype(int)

df1 = df1.assign(Type= df1['Type'].map({'A': 1, 'B': 2, 'C': 3}))
df1.head(10)
```

Now let's have a look at the statistics of our dataset

```python
df1.describe()
```

We can draw some interesting observation here.
1. Negative sales: Depending on the retailers, Weekly_Sales here might refer to Net Sales (=Gross Sales - Return), which can be negative. However, I'd still check with the retailer to make sure the data is accurate here.
2. Negative markdown in MarkDown 2 and Markdown 3 colums: Again, I'd confirm with the retailer to ensure high quality data.

Since we have 5 different types of MarkDown, I created a total MarkDown column to see the aggregated MarkDown effect on Weekly_Sales

```python
df1['Total_Markdown'] = df1['MarkDown1'] + df1['MarkDown2'] + df1['MarkDown3'] + df1['MarkDown4'] + df1['MarkDown5']
```

### Part 2: Exploratory data analysis

Let's first have a look at the Weekly_Sales pattern over time.

```python
df_Sales = df1.groupby(by=['Date'], as_index=False)['Weekly_Sales'].sum()

plt.rcParams['figure.figsize'] = [16,4]
plt.plot(df_Sales.Date, df_Sales.Weekly_Sales)
plt.show()
```

Looking at the graph above, we can clearly see:
- **Seasonality:** There's a repeated pattern in Weekly_Sales over time. Sales are typically stable throughout the year, and then have two big spikes in December. This implies that next year's sales can be roughly predicted by looking at previous year' sales.
- **Trend:** There is no noticeable upward or downward trend in the data. This is also because we only have data for less than 3 years, which is quite limited
- **Noise:** Even though there's a seasonality effect in Sales, we can see there is still a lot of random variation. This can make it difficult to predict exact next year's Sales.

Given the big seasonality effect in sale figures, let's try to analyze it further. Do all stores and departments closely follow the overall Sales patterns, or it is due to a few stores/departments that cause the sale spikes?

### 1. Department analysis
We will analyze the top performing departments and their contribution to the overall sales

```python
dept_most_sales = df1.groupby('Dept')['Weekly_Sales'].sum().nlargest(10)
dept_most_sales
```

```python
df_Dept = df1.set_index('Dept')
df_Dept = df_Dept.loc[dept_most_sales.index]

plt.rcParams['figure.figsize'] = [16,4]
df_Dept.pivot_table(index='Date', columns='Dept', values='Weekly_Sales').plot()
plt.show()
```

In turns out that department 72 was the main cause of Decamber sale spikes. Its sale was relevatively stable throughout the year, yet skyrocketed during December each year.

On the other hand, other departments' sales were more stable throughout the year. 

```python
df_Dept_top2 = df_Dept.loc[dept_most_sales.head(2).index]

df_Dept_top2.pivot_table(index='Date', columns='Dept', values='Weekly_Sales').plot()
plt.show()
```

We can also see that the top 2 departments, 92 and 95, have opposite sale cycles. When sales in one department go up, the other one go down and vice versa. Since these two departments have a significant contribution to the total sales, the retailer can use this information to stablize sales and manage seasonal fluctuations. 

Since December is known for promotions and markdowns, let's check if department 72's sale spikes were due to promotional markdowns.

```python
df_Dept_72 = df1[df1.Dept == 72]
df_Dept_72 = df_Dept_72.groupby(by=['Date'], as_index=False)['Weekly_Sales', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'].sum()

df_Dept_72.plot(x='Date', y=['Weekly_Sales', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']);
plt.show()
```

 We can see that MarkDown3 was the most effective to trigger sales in Department 72. MarkDown2 also had some impact, even though its occurence lagged a bit behind the second spike.

With this information, the retailer can know that certain MarkDown is most effective on certain department, and therefore, will be able to create more effective promotion campaigns.

### 2. Store analysis
In this section, we will analyze the top performing stores and factors impacting store sales

```python
stores.set_index('Store', inplace=True, drop=True)
store_most_sales = df1.groupby('Store')['Weekly_Sales'].sum().nlargest(10)
store_most_sales.index

stores.loc[store_most_sales.index]
```

Looking at the 10 stores with highest sales, it's very interesting that 90% of them are type A

### 3. Markdown analysis
In this part, we will have a closer look on the effect of markdowns on overall sales

```python
df_Markdown = df1.groupby(by=['Date'], as_index=False)['Weekly_Sales', 'Total_Markdown'].sum()

df_Markdown.Weekly_Sales.plot(legend=True);
df_Markdown.Total_Markdown.plot(secondary_y=True, legend=True);
plt.show()
```

We see that Markdown does not have that much of a big effect of Week_Sales. Before making any conclusion, let's try to break  it down to each Markdown type.

```python
def graph(df, markdown):
    new_df = df.groupby(by=['Date'], as_index=False)['Weekly_Sales', markdown].sum()
    new_df.plot(x='Date', y=['Weekly_Sales', markdown]);
    plt.show()

markdown_list = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

for x in markdown_list:
    graph(df1, x)
```

Among all the Markdown, MarkDown5 has the most impact on sales. 

We saw that Markdown doesn't have a big contribution to sales. It'd be very interesting to see the optimal range of Markdown the retailer should offer to customers

```python
def graph_2(df, markdown):
    df.plot.scatter(x=markdown, y='Weekly_Sales');

markdown_list = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

for x in markdown_list:
    graph_2(df1, x)
```

Looking at the graphs, the ideal range for all MarkDowns, except for MarkDown 3, is between 0 to 10000. For MarkDown3, its ideal range falls into (0, 5000) and (60000,110000). 

### 4. Sales seasonality
We will attempt to see how sales are distributed throughout the year

```python
plt.rcParams['figure.figsize'] = [10,5]
df1['Month'] = df1['Date'].dt.month

df1.plot.scatter(x='Month', y="Weekly_Sales");
```

The graph shows that in general, sales are relatively stable throughout the year, except for November and December.

Let's also see how MarkDown is distributed

```python
df1.plot.scatter(x='Month', y="Total_Markdown");
```

This once again confirmed our previous observations: Markdown campaigns are not very effective. For instance, the big markdown in February did not result in higher sales

Now instead of the distribution, let's check the aggregate sales throughout the year.
But first, we need to select only a subset of the dataframe, from 2010/02/05 to 2012/02/03. Since we have approximately 2.5 years of data, we need to make sure all months have the same number of data points.

```python
df2 = df1.set_index('Date')
df2.sort_index
df2 = df2.loc['2010-02-05':'2012-02-03']
df2.head(10)
```

```python
df_Month = df2.groupby(by=['Month'], as_index=False)['Weekly_Sales'].sum()

plt.rcParams['figure.figsize'] = [10,5]
df_Month.plot.bar(x = 'Month', y = 'Weekly_Sales', rot=0)
plt.show()
```

February, April, July and December are the months generating the most sales

However, we have big holidays, such as Christmas fall into those mentioned months. Let's see what happens if we exclude the Holiday week, since sales in Holiday weeks are higher than sales in normal weeks.

```python
df1.groupby('IsHoliday')['Weekly_Sales'].mean().plot.barh();
```

```python
holiday_week = df2.loc[df2['IsHoliday'] == True]
normal_week = df2.loc[df2['IsHoliday'] == False]

df_Month_3 = normal_week.groupby(by=['Month'], as_index=False)['Weekly_Sales'].sum()

df_Month_3.plot.bar(x = 'Month', y = 'Weekly_Sales', rot=0)
plt.show()
```

```python
df_Month_4 = holiday_week.groupby(by=['Month'], as_index=False)['Weekly_Sales'].sum()

df_Month_4.plot.bar(x = 'Month', y = 'Weekly_Sales', rot=0)
plt.show()
```

Excluding holiday weeks, April, July and December still remain the top 3 months with most sales. However, as holiday weeks have a big contribution to February sales, February organic sales are lower than other months.
